# 1.Import Liabraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 2.Load and Explore the Data

```python
data = pd.read_csv("Sample - Superstore.csv" , encoding = 'latin-1')
print(data)
```

```python
# display first few rows
print(data.head())
```

```python
# Data overview
print(data.info())
```

```python
print(data.describe())
```

```python
# Check for null values
print(data.isnull().sum())
```

# 3.Data Cleaning

```python
# Remove Duplicates
print(data.drop_duplicates(inplace=True))
```

```python
# Convert Order date to DateTime Format
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
print(data)
```

# 4.Key Analysis

```python
data['Month'] = data['Order Date'].dt.month
Monthly_sales = data.groupby(['Month']).agg({'Sales':'sum'}).squeeze()
plt.figure(figsize=(10,5))
plt.plot(Monthly_sales.index , Monthly_sales.values , marker='o' , color = 'hotpink')
plt.title('Monthly Sales Trennd')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid()
plt.show()
```

```python
# identify the month with peak sales
peak_sales_month = Monthly_sales.idxmax()
peak_sales_values = Monthly_sales.max()
print(f"The month with peak sales is {peak_sales_month} with total sales of $ {peak_sales_values:.2f}")
```

```python
region_sales = data.groupby(['Region']).agg({'Sales':'sum'}).squeeze()
region_profit = data.groupby(['Region']).agg({'Profit':'sum'}).squeeze()

# Plot Sales and Profit by Region
region_sales.plot(kind='bar' , color = 'skyblue' , alpha = 0.7 , label = 'Sales' , figsize=(8,5))
region_profit.plot(kind='bar' , color = 'orange' , alpha = 0.7 , label = 'Profit')
plt.title('Sales and Profit by Region')
plt.xlabel('Region')
plt.ylabel('Amount')
plt.legend()
plt.show()
```

```python
# Region with Maximum sales and Profit
top_region_sales = region_sales.idxmax()
top_region_sales_values = region_sales.max()
top_region_profit = region_profit.idxmax()
top_region_profit_values = region_profit.max()
print(f"The Region with Highest Sales {top_region_sales} with total sales of $ {top_region_sales_values:.2f}")
print(f"The Region with Highest Profit {top_region_profit} with total profit of $ {top_region_profit_values:.2f}")

```

```python
category_sales = data.groupby(['Category']).agg({'Sales':'sum'}).squeeze()
category_profit = data.groupby(['Category']).agg({'Profit':'sum'}).squeeze()
category_sales.plot(kind = 'bar' , color = 'green' , alpha = 0.8 , figsize=(8,5))
category_profit.plot(kind = 'bar' , color = 'red' , alpha = 0.8)
plt.legend()
plt.title('Sales and Profit by Category')
plt.xlabel('Category')
plt.ylabel('Amount')
plt.show()
```

```python
# Category with Maximum sales and Profit
top_category_sales = category_sales.idxmax()
top_category_sales_values = category_sales.max()
top_category_profit = category_profit.idxmax()
top_category_profit_values = category_profit.max()
print(f"The category with Highest Sales {top_category_sales} with total sales of $ {top_category_sales_values:.2f}")
print(f"The category with Highest Profit {top_category_profit} with total profit of $ {top_category_profit_values:.2f}")

```

```python
top_subcategories = data.groupby(['Sub-Category']).agg({'Sales':'sum'}).squeeze()
top_subcategories.plot(kind='bar' , color = 'purple' , alpha = 0.8 , figsize=(8,5))
plt.title('Sub-Category by Sales')
plt.xlabel('Sub-Category')
plt.ylabel('Amount')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(data['Discount'] , data['Profit'] , alpha = 0.5 , color = 'magenta')
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.grid()
plt.show()
```

```python
correlation = data['Discount'].corr(data['Profit'])
print(f"The correlation between Discount and Profit is: {correlation:.2f}")
```

```python
Segment_sales = data.groupby('Segment')['Sales'].sum()
Segment_sales.plot(kind = 'pie' , autopct = '%1.1f%%' , startangle = 90 , colors = ['gold' , 'lightblue' , 'lightgreen'])
plt.title('Sales Contribution by Segment')
plt.show()
```
